```python
from random import randrange

import numpy as np
import time
```

```python
class Agent:
    a = 0;
    
class Item:
    a = 0;
    
class CentralRegister:
    agents = []
    items = []
    itemsRegister = []
    
    def init(self):
        self.agents = []
        self.items = []
        self.itemsRegister = []
    
    def addAgent(self, agent):
        self.agents.append(agent);
        
        for item in self.itemsRegister:
            item.append(0);
        
        return
            
    def removeAgent(self, agent):
        index = self.agents.index(agent)
        self.agents.pop(index)
        
        for item in self.itemsRegister:
            item.pop(index)
        
        return
        
    def setItemQuantity(self, receiver, itemIndex, quantity):
        receiverIndex = self.agents.index(receiver)
        entry = self.itemsRegister[itemIndex]
        entry[receiverIndex] += quantity
        
        return
        
    def transferItem(self, sender, receiver, itemIndex, quantity):
        self.setItemQuantity(sender, itemIndex, -quantity)
        self.setItemQuantity(receiver, itemIndex, quantity)
        
        return
        
    def addNewItemType(self, item):
        self.items.append(item)
        
        newEntry = []
        for agents in self.agents:
            newEntry.append(0)
            
        self.itemsRegister.append(newEntry)

        return
    
    def getAgentHoldingDistro(self, agent):
        holdings = []
        agentIndex = self.agents.index(agent)
    
        for item in self.itemsRegister:
            holdings.append(item[agentIndex])
            
        return holdings
    
    def getAgentItemHolding(self, agent, itemIndex):
        agentIndex = self.agents.index(agent)
        
        return self.itemsRegister[itemIndex][agentIndex]
    
    def getItemDistribution(self, item):
        itemIndex = self.items.index[item]
        
        return self.itemsRegister[itemIndex]
    
    def getItemList(self):
        return self.items
    
    def getAgentsList(self):
        return self.agents
    
    def getItemsRegister(self):
        return self.itemsRegister
    
    def getItemTotQnt(self, itemIndex):
        distro = self.itemsRegister[itemIndex]
        totQnt = 0
        
        for i in range(len(distro)):
            totQnt += distro[i]
            
        return totQnt
        
```

```python
class Model:
    
    centralRegister = None
    time = 0
    
    def init(self):
        
        self.centralRegister = CentralRegister()
        self.centralRegister.init()
        #print(self.centralRegister.getItemsRegister())
        
        self.centralRegister.addNewItemType(Item())
        #print(self.centralRegister.getItemsRegister())
        
        for i in range(1000):
            newAgent = Agent()
            self.centralRegister.addAgent(newAgent)
            self.centralRegister.setItemQuantity(newAgent, 0, 10)
            
        #print(self.centralRegister.getItemsRegister())
        
        return
    
    def step(self):
        
        agents = self.centralRegister.getAgentsList();
        
        for i in range(len(agents)):
            agentCash = self.centralRegister.getAgentItemHolding(agents[i], 0)
            qntExchanged = randrange(agentCash)
            receiverAgentIndex = randrange(len(agents))
            
            self.centralRegister.transferItem(agents[i], agents[receiverAgentIndex], 0, qntExchanged)
            
            self.time += 1
            
        if (self.time % 1 == 10000):
            print(self.centralRegister.getItemsRegister())
            print(self.centralRegister.getItemTotQnt(0))
t0All = time.time()    
```

```python
t0All = time.time()

model = Model()

model.init()

for i in range(1000):
    model.step()

t1All = time.time()
elapsed = np.round(t1All - t0All, 1)
print('Total elapsed time:                ' + str(elapsed) + 's')

```
